#  Parkinson’s Disease Dataset Information  

##  Overview  
This dataset consists of **195 entries (rows) and 24 attributes (columns)**. It contains various voice measurements used to detect **Parkinson’s disease**. The dataset includes **one categorical feature (`status`)** and **multiple numerical features**, all of which describe different frequency and amplitude variations in the voice.  

## 🔹 Data Details  
- The dataset has **no missing values**, as every column contains **195 non-null** entries.  
- The **majority of the columns (22 out of 24)** are of type `float64`, representing various frequency-related measurements.  
- The `status` column is an `int64` type, acting as a label to indicate whether a person **has Parkinson’s disease (1) or not (0)**.  
- The `name` column is the only `object` type feature, containing the names of individuals.  

##  Feature Descriptions  
1. **MDVP:Fo(Hz), MDVP:Fhi(Hz), MDVP:Flo(Hz)** → Fundamental frequency and its variations in the voice.  
2. **Jitter Features (MDVP:Jitter(%), MDVP:Jitter(Abs), MDVP:RAP, MDVP:PPQ, Jitter:DDP)** → Measure variations in pitch.  
3. **Shimmer Features (MDVP:Shimmer, MDVP:Shimmer(dB), Shimmer:APQ3, Shimmer:APQ5, MDVP:APQ, Shimmer:DDA)** → Measure variations in amplitude.  
4. **NHR & HNR** → Noise-to-Harmonics and Harmonics-to-Noise ratios, used to evaluate voice quality.  
5. **RPDE & DFA** → Measures of signal dynamics and fractal scaling exponent.  
6. **Spread1 & Spread2** → Indicators of signal deviations.  
7. **D2 & PPE** → Nonlinear dynamical complexity measures of the voice signal.  

## 🏷 Target Variable  
- **`status`**: The key column indicating Parkinson’s disease diagnosis:  
  - `1` → Person **has** Parkinson’s disease.  
  - `0` → Person **does not have** Parkinson’s disease.

# Data visualization

## Univariate Analysis

```python
def distplots(df):
    num_cols = list(df.columns)[1:]  # Exclude first column if it's non-numeric
    num_features = len(num_cols)
    
    rows = (num_features // 3) + (num_features % 3 > 0)  # Calculate required rows
    fig, axes = plt.subplots(rows, 3, figsize=(18, 4 * rows))  # Adjust figure size
    
    # Flatten axes array to easily iterate
    axes = axes.flatten()

    for i, col in enumerate(num_cols):
        sns.histplot(df[col], ax=axes[i], kde=True)  # KDE for better visualization
        axes[i].set_title(f"Distribution of {col}")
    
    # Remove empty subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

distplots(df)
```

### **Univariate Analysis: Key Insights**

From the **distribution plots**, we can observe the following patterns in the dataset:

1. **Skewness in Frequency-Based Features:**  
   - Features like `MDVP:Fo(Hz)`, `MDVP:Fhi(Hz)`, and `MDVP:Flo(Hz)` exhibit right-skewed distributions, indicating that most individuals have lower fundamental frequency values, while a few have significantly higher values.

2. **Jitter and Shimmer Variability:**  
   - `MDVP:Jitter(%)`, `MDVP:Jitter(Abs)`, `MDVP:RAP`, and `Jitter:DDP` show strong right-skewness, suggesting that the majority of subjects have low jitter (frequency perturbation), while a few have much higher values.  
   - Similarly, shimmer-related features (`MDVP:Shimmer`, `MDVP:Shimmer(dB)`, `Shimmer:APQ3`, etc.) also exhibit right-skewed distributions, indicating variations in amplitude perturbation.

3. **Harmonic-to-Noise Ratio (HNR):**  
   - `HNR` appears to have a more **normally distributed** shape, meaning most individuals have a balanced ratio of harmonic to noise energy.

4. **Parkinson's Disease Status:**  
   - The `status` feature is **bimodal**, meaning it has two distinct peaks, which likely correspond to **healthy individuals (0)** and **Parkinson’s patients (1)**.

5. **Complexity Features (`RPDE`, `DFA`, `spread1`, `spread2`, `D2`, `PPE`)**  
   - Features such as `RPDE`, `DFA`, and `spread1` are **approximately normal**, which suggests that these measures of signal complexity may be more stable across the population.

### Next Steps:
- **Feature Scaling:** Many of the features are highly skewed, so transformations like **log-scaling** or **Box-Cox transformation** may be useful for improving model performance.
- **Outlier Detection:** Box plots (not shown here) can help confirm the presence of **outliers**, particularly in `Jitter` and `Shimmer` features.
- **Class Balance Analysis:** The **bimodal distribution** in `status` suggests an imbalanced dataset, which should be accounted for in machine learning models.

```python
# Function to create box plots in a 3-column layout
def boxplots(df):
    num_cols = list(df.columns)[1:]  # Exclude first column if it's non-numeric
    num_features = len(num_cols)
    
    # Set up the figure with multiple subplots (rows adjusted based on features)
    rows = (num_features // 3) + (num_features % 3 > 0)  # Calculate required rows
    fig, axes = plt.subplots(rows, 3, figsize=(18, 4 * rows))  # Adjust figure size
    
    # Flatten axes array to easily iterate
    axes = axes.flatten()

    for i, col in enumerate(num_cols):
        sns.boxplot(y=df[col], ax=axes[i])  # Vertical box plot
        axes[i].set_title(f"Boxplot of {col}")
    
    # Remove empty subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

boxplots(df)
```

###  **Boxplot Analysis: Key Insights**  

The boxplots provide valuable information about the **distribution, spread, and presence of outliers** in the dataset:

1. **Presence of Outliers:**  
   - Several features, especially `MDVP:Fhi(Hz)`, `MDVP:Jitter(%)`, `MDVP:Shimmer(dB)`, `NHR`, and `PPE`, have **many outliers**.  
   - This suggests that some patients exhibit **significant variations in vocal attributes**, likely indicating the effects of Parkinson’s disease.

2. **Asymmetry in Distributions:**  
   - Many features such as `MDVP:Jitter(%)`, `MDVP:Shimmer`, and `NHR` show **right-skewed distributions**, meaning most values are concentrated at the lower end, while a few extreme values exist at the higher end.
   - `HNR`, on the other hand, has **a few low-end outliers**, which indicates some individuals have unusually poor harmonic-to-noise ratios.

3. **Feature Variability:**  
   - Some features, like `spread1` and `DFA`, have a more **compact interquartile range (IQR)**, suggesting less variation in these features.  
   - Others, such as `MDVP:Fhi(Hz)` and `PPE`, show **wider IQRs**, indicating a larger spread of values.

# Data Preprocessing

## Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler((-1,1))
x = x_smote
x = scaler.fit_transform(x)
y = y_smote
```

```python
import sklearn
print(sklearn.__version__)
```

```python
print("Columns in X:", x_smote.columns)
```

## PCA

```python
from sklearn.decomposition import PCA

pca  = PCA(.99)
x_pca = pca.fit_transform(x)
print(x.shape)
print(x_pca.shape)
```

```python
pd.DataFrame(x_pca)
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state = 1)
```

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, BernoulliNB
```

```python
# Define models
models = {
    "Logistic Regression": LogisticRegression(C=0.4, max_iter=1000, solver='liblinear'),
    "Decision Tree": DecisionTreeClassifier(random_state=14),
    "Random Forest": RandomForestClassifier(random_state=14, criterion='entropy'),
    "Support Vector Machine": SVC(probability=True),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "GaussianNB": GaussianNB(),
    "BernoulliNB": BernoulliNB()
}

# Dictionary to store accuracy scores
model_accuracies = {}

# Train and evaluate each model
for name, model in models.items():
    model.fit(x_train, y_train)  # Train model
    y_pred = model.predict(x_test)  # Predict
    
    # Compute accuracy
    accuracy = accuracy_score(y_test, y_pred)
    model_accuracies[name] = accuracy

    # Print evaluation metrics
    print(f"\n{name} Accuracy: {accuracy:.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
voting_clf = VotingClassifier(
    estimators=[
        ('log_reg', models["Logistic Regression"]),
        ('dt', models["Decision Tree"]),
        ('rf', models["Random Forest"]),
        ('svm', models["Support Vector Machine"]),
        ('knn', models["K-Nearest Neighbors"]),
        ('gnb', models["GaussianNB"]),
        ('bnb', models["BernoulliNB"])
    ],
    voting='hard'
)

# Train Voting Classifier
voting_clf.fit(x_train, y_train)

# Evaluate Voting Classifier
y_pred_voting = voting_clf.predict(x_test)
accuracy_voting = accuracy_score(y_test, y_pred_voting)
model_accuracies["Voting Classifier"] = accuracy_voting

print("\nVoting Classifier Accuracy:", accuracy_voting)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_voting))
print("Classification Report:\n", classification_report(y_test, y_pred_voting))
```

```python
plt.figure(figsize=(10, 6))
plt.barh(list(model_accuracies.keys()), list(model_accuracies.values()), color='skyblue')
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.title('Model Performance Comparison')
plt.xlim(0, 1)  # Accuracy ranges from 0 to 1
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show accuracy values on bars
for index, value in enumerate(model_accuracies.values()):
    plt.text(value, index, f"{value:.4f}", va='center')

plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Get the predicted probabilities for the positive class (1)
y_probs_rf = models["K-Nearest Neighbors"].predict_proba(x_test)[:, 1]

# Compute ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_probs_rf)

# Compute AUC Score
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'Random Forest (AUC = {roc_auc:.4f})')

# Plot Diagonal Line (Random Guessing)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')

# Labels and Title
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for K-Nearest Neighbors')
plt.legend(loc="lower right")

# Show Plot
plt.show()
```

### **ROC Curve Analysis for K-Nearest Neighbors (KNN)**  

1. **Model Performance:**  
   - The **AUC (Area Under Curve) value is 0.9971**, which is **exceptionally high**.
   - This indicates that the **K-Nearest Neighbors (KNN) model is performing very well** in distinguishing between positive and negative classes.

2. **ROC Curve Shape:**  
   - The curve **almost reaches the top-left corner (TPR = 1, FPR = 0)**, suggesting that the model has **very high sensitivity (recall)** and is capable of detecting true positives effectively.
   - A model with a perfect AUC score of 1 would form a right-angle shape, and the current curve is very close to that.

3. **Comparison to Random Guessing:**  
   - The **red dashed diagonal line** represents a random classifier with **AUC = 0.5**.
   - Since our **AUC is significantly higher (0.9971)**, it confirms that the model has strong predictive capabilities.

### ** Conclusion:**  
The **K-Nearest Neighbors model is highly effective** in predicting the target variable.  
**Low false positives & high true positives** make it an excellent model choice.  

## Training a Neural Netwrok

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
# Define the model
model = Sequential([
    Dense(128, activation='relu', input_shape=(x_pca.shape[1],)),  # Input Layer
    Dropout(0.2),  # Regularization to prevent overfitting
    Dense(64, activation='relu'),  # Hidden Layer
    Dropout(0.2),
    Dense(32, activation='relu'),  # Hidden Layer
    Dense(1, activation='sigmoid')  # Output Layer for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Model Summary
model.summary()
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=50, batch_size=16, validation_data=(x_test, y_test))
```

## Neural Netwrok Model Evaluation

```python
# Evaluate on Test Data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
# Plot Accuracy and Loss
plt.figure(figsize=(12, 4))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training vs. Validation Accuracy')

# Loss Plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training vs. Validation Loss')

plt.show()
```
