# Data Preprocessing

## Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler((-1,1))
x = x_smote
x = scaler.fit_transform(x)
y = y_smote
```

```python
import sklearn
print(sklearn.__version__)
```

```python
print("Columns in X:", x_smote.columns)
```

## PCA

```python
from sklearn.decomposition import PCA

pca  = PCA(.99)
x_pca = pca.fit_transform(x)
print(x.shape)
print(x_pca.shape)
```

```python
pd.DataFrame(x_pca)
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state = 1)
```

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, BernoulliNB
```

```python
# Define models
models = {
    "Logistic Regression": LogisticRegression(C=0.4, max_iter=1000, solver='liblinear'),
    "Decision Tree": DecisionTreeClassifier(random_state=14),
    "Random Forest": RandomForestClassifier(random_state=14, criterion='entropy'),
    "Support Vector Machine": SVC(probability=True),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "GaussianNB": GaussianNB(),
    "BernoulliNB": BernoulliNB()
}

# Dictionary to store accuracy scores
model_accuracies = {}

# Train and evaluate each model
for name, model in models.items():
    model.fit(x_train, y_train)  # Train model
    y_pred = model.predict(x_test)  # Predict
    
    # Compute accuracy
    accuracy = accuracy_score(y_test, y_pred)
    model_accuracies[name] = accuracy

    # Print evaluation metrics
    print(f"\n{name} Accuracy: {accuracy:.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
voting_clf = VotingClassifier(
    estimators=[
        ('log_reg', models["Logistic Regression"]),
        ('dt', models["Decision Tree"]),
        ('rf', models["Random Forest"]),
        ('svm', models["Support Vector Machine"]),
        ('knn', models["K-Nearest Neighbors"]),
        ('gnb', models["GaussianNB"]),
        ('bnb', models["BernoulliNB"])
    ],
    voting='hard'
)

# Train Voting Classifier
voting_clf.fit(x_train, y_train)

# Evaluate Voting Classifier
y_pred_voting = voting_clf.predict(x_test)
accuracy_voting = accuracy_score(y_test, y_pred_voting)
model_accuracies["Voting Classifier"] = accuracy_voting

print("\nVoting Classifier Accuracy:", accuracy_voting)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_voting))
print("Classification Report:\n", classification_report(y_test, y_pred_voting))
```

```python
plt.figure(figsize=(10, 6))
plt.barh(list(model_accuracies.keys()), list(model_accuracies.values()), color='skyblue')
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.title('Model Performance Comparison')
plt.xlim(0, 1)  # Accuracy ranges from 0 to 1
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show accuracy values on bars
for index, value in enumerate(model_accuracies.values()):
    plt.text(value, index, f"{value:.4f}", va='center')

plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Get the predicted probabilities for the positive class (1)
y_probs_rf = models["K-Nearest Neighbors"].predict_proba(x_test)[:, 1]

# Compute ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_probs_rf)

# Compute AUC Score
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'Random Forest (AUC = {roc_auc:.4f})')

# Plot Diagonal Line (Random Guessing)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')

# Labels and Title
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for K-Nearest Neighbors')
plt.legend(loc="lower right")

# Show Plot
plt.show()
```

## Training a Neural Netwrok

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
# Define the model
model = Sequential([
    Dense(128, activation='relu', input_shape=(x_pca.shape[1],)),  # Input Layer
    Dropout(0.2),  # Regularization to prevent overfitting
    Dense(64, activation='relu'),  # Hidden Layer
    Dropout(0.2),
    Dense(32, activation='relu'),  # Hidden Layer
    Dense(1, activation='sigmoid')  # Output Layer for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Model Summary
model.summary()
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=50, batch_size=16, validation_data=(x_test, y_test))
```

## Neural Netwrok Model Evaluation

```python
# Evaluate on Test Data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
# Plot Accuracy and Loss
plt.figure(figsize=(12, 4))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training vs. Validation Accuracy')

# Loss Plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training vs. Validation Loss')

plt.show()
```
